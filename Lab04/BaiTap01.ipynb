```python
import pandas as pd

X = pd.read_csv('housing_RT.csv').drop(columns=['id'])
X.head()
```

```python
import numpy as np
price_mean = np.mean(X.price)
print(f'Mean Price: {price_mean}')
```

```python
y = [1 if price < price_mean else 0 for price in X.price]
y = np.array(y)
y
```

```python
import numpy as np
def my_perceptron(X, y, eta, lanlap):
    n = len(X[0, ])
    m = len(X[:, 0])
    print(f'm = {m}\nn = {n}')
    w0 = -0.2
    w = np.random.randn(n)
    print(f'w0 = {w0}\nw = {w}')
    for t in range(0, lanlap):
        print('lanlap_____', t + 1)
        for i in range(0, m):
            gx = w0 + sum(X[i, ] * w)
            print(f'gx = {gx}')
            if gx > 0:
                output = 1
            else:
                output = 0
            w0 = w0 + eta * (y[i] - output) * 1
            w = w + eta * (y[i] - output) * X[i, ]
            print(f'w0 = {w0}\nw = {w}')
    return (np.round(w0, 3), np.round(w, 3))
```

```python
w0, w = my_perceptron(X.to_numpy(), y, 0.15, 2)
```

```python
print(f'W0 = {w0}')
for i in range(0, len(w)):
    print(f'W{i + 1} = {w[i]}')
```
